```python
records = []

if __name__ == '__main__':
    for _ in range(int(input())):
        name = input()
        score = float(input())
        records.append([name, score])



    scores = [score for name, score in records]
    sortrec = list(set(scores))
    sortrec.sort(reverse = True)
    seconds = [name for name , score in records if score == sortrec[1]]
    print(seconds)
```
